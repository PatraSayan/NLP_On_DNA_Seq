```python
train["Clean_Seq"] = train["NucleotideSequence"].str.replace(r"[<>]","", regex=True)
vald["Clean_Seq"] = vald["NucleotideSequence"].str.replace(r"[<>]","", regex = True)
test["Clean_Seq"] = test["NucleotideSequence"].str.replace(r"[<>]", "", regex=True)
```

```python
train
```

```python
train["Sequence_length"] = train["Clean_Seq"].apply(len)
vald["Sequence_length"] = vald["Clean_Seq"].apply(len)
test["Sequence_length"] = test["Clean_Seq"].apply(len)
```

```python
fig, axes = plt.subplots(3,1, figsize = (10,10))

sns.histplot(data = train["Sequence_length"], color= "#268df8", bins=25, ax=axes[0])
axes[0].set_ylabel("length of Nucleotide")
axes[0].set_xlabel("")
axes[0].set_title("Train Data Nucleotide Length")

sns.histplot(data = vald["Sequence_length"], color= "#cc26f8", bins=25, ax=axes[1])
axes[1].set_ylabel("length of Nucleotide")
axes[1].set_xlabel("")
axes[1].set_title("Validation Data Nucleotide Length")

sns.histplot(data = test["Sequence_length"], color= "#ff385e", bins=25, ax=axes[2])
axes[2].set_ylabel("length of Nucleotide")
axes[2].set_xlabel("")
axes[2].set_title("Test Data Nucleotide Length")
```

```python
fig, axes = plt.subplots(3,1, figsize = (14,14))

sns.boxplot(data = train, x = "GeneType", y = "Sequence_length", hue = "GeneType", palette = "cubehelix",order = sorted(train["GeneType"].unique()) ,ax = axes[0])
axes[0].set_title("Train Data")
axes[0].set_xlabel("")

sns.boxplot(data = vald, x = "GeneType", y = "Sequence_length", hue = "GeneType", palette = "cubehelix",order = sorted(vald["GeneType"].unique()) ,ax = axes[1])
axes[1].set_title("Validation Data")
axes[1].set_xlabel("")

sns.boxplot(data = test, x = "GeneType", y = "Sequence_length", hue = "GeneType", palette = "cubehelix",order = sorted(test["GeneType"].unique()) ,ax = axes[2])
axes[2].set_title("Test Data")
axes[2].set_xlabel("");
```

```python
def nucleotide_count(seq):
    A = seq.count("A")
    T = seq.count("T")
    G = seq.count("G")
    C = seq.count("C")
    return pd.Series({
        "A": A,
        "T": T,
        "G": G,
        "C": C})
```

```python
train_nt_counts = train["Clean_Seq"].apply(nucleotide_count)
train = pd.concat([train, train_nt_counts], axis=1)
```

```python
vald_nt_counts = vald["Clean_Seq"].apply(nucleotide_count)
vald = pd.concat([vald, vald_nt_counts], axis=1)
```

```python
test_nt_counts = test["Clean_Seq"].apply(nucleotide_count)
test = pd.concat([test, test_nt_counts], axis=1)
```

```python
fig, axes = plt.subplots(3, 1, figsize = (14,10))

train.groupby("GeneType")
```

```python
fig, axes = plt.subplots(3, 1, figsize = (14, 14))

train.groupby("GeneType")[['A', 'T', 'G', 'C']].mean().plot(kind = "bar", stacked = True,ax = axes[0])
axes[0].tick_params(axis='x', rotation=0)

vald.groupby("GeneType")[['A', 'T', 'G', 'C']].mean().plot(kind = "bar", stacked = True,ax = axes[1])
axes[1].tick_params(axis='x', rotation=0)

test.groupby("GeneType")[['A', 'T', 'G', 'C']].mean().plot(kind = "bar", stacked = True,ax = axes[2])
axes[2].tick_params(axis='x', rotation=0)
```

```python
train["GC_content"] = ((train["G"] + train["C"]) / train["Sequence_length"])*100
vald["GC_content"] = ((vald["G"] + vald["C"]) / vald["Sequence_length"])*100
test["GC_content"] = ((test["G"] + test["C"]) / test["Sequence_length"])*100
```

```python
fig, axes = plt.subplots(3,1, figsize = (10,10))

sns.histplot(data = train["GC_content"], color= "#268df8", bins=25, ax=axes[0])
axes[0].set_ylabel("count")
axes[0].set_xlabel("")
axes[0].set_title("Train Data GC Content Distribution")

sns.histplot(data = vald["GC_content"], color= "#cc26f8", bins=25, ax=axes[1])
axes[1].set_ylabel("count")
axes[1].set_xlabel("")
axes[1].set_title("Validation Data GC Content Distribution")

sns.histplot(data = test["GC_content"], color= "#ff385e", bins=25, ax=axes[2])
axes[2].set_ylabel("count")
axes[2].set_xlabel("")
axes[2].set_title("Test Data GC Content Distribution")
```

```python
fig, axes = plt.subplots(3,1, figsize = (14,14))

sns.boxplot(data = train, x = "GeneType", y = "GC_content", hue = "GeneType", palette = "cubehelix",order = sorted(train["GeneType"].unique()) ,ax = axes[0])
axes[0].set_title("Train Data GC Count By Gene Type")
axes[0].set_xlabel("")

sns.boxplot(data = vald, x = "GeneType", y = "GC_content", hue = "GeneType", palette = "cubehelix",order = sorted(vald["GeneType"].unique()) ,ax = axes[1])
axes[1].set_title("Validation Data GC Count By Gene Type")
axes[1].set_xlabel("")

sns.boxplot(data = test, x = "GeneType", y = "GC_content", hue = "GeneType", palette = "cubehelix",order = sorted(test["GeneType"].unique()) ,ax = axes[2])
axes[2].set_title("Test Data GC Count By Gene Type")
axes[2].set_xlabel("");
```

```python
train_correlation = train[["A", "T", "G", "C", "GC_content"]].corr()
vald_correlation = vald[["A", "T", "G", "C", "GC_content"]].corr()
test_correlation = test[["A", "T", "G", "C", "GC_content"]].corr()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15,4))

sns.heatmap(train_correlation, annot=True, cmap = "icefire", ax=axes[0])
axes[0].set_title("Train Correlation")

sns.heatmap(vald_correlation, annot=True, cmap = "icefire", ax=axes[1])
axes[1].set_title("Train Correlation")

sns.heatmap(test_correlation, annot=True, cmap = "icefire", ax=axes[2])
axes[2].set_title("Train Correlation");
```

```python
from wordcloud import WordCloud
train_text = ' '.join(train["Description"].astype(str))
traincloud = WordCloud(width=800, height=400, background_color="white").generate(train_text)

vald_text = ' '.join(vald["Description"].astype(str))
valdcloud = WordCloud(width=800, height=400, background_color="white").generate(vald_text)

test_text = ' '.join(test["Description"].astype(str))
testcloud = WordCloud(width=800, height=400, background_color="white").generate(test_text)
```

```python
fig, axes = plt.subplots(3, 1, figsize = (12, 20))

axes[0].imshow(traincloud, interpolation="bilinear")
axes[0].axis('off')
axes[0].set_title("Train Data", fontsize = 20, fontweight = "bold");

axes[1].imshow(valdcloud, interpolation="bilinear")
axes[1].axis('off')
axes[1].set_title("Validation Data", fontsize = 20, fontweight = "bold");

axes[2].imshow(testcloud, interpolation="bilinear")
axes[2].axis('off')
axes[2].set_title("Test Data", fontsize = 20, fontweight = "bold");
```

```python
train_freq = train[['A','T','G','C']].sum()
vald_freq = vald[['A','T','G','C']].sum()
test_freq = test[['A','T','G','C']].sum()

fig, axes = plt.subplots(1, 3, figsize = (20,8))
train_freq.plot(kind = "pie", autopct ="%1.1f%%", ax = axes[0])
axes[0].set_title("Train Overall Nucleotide Frequency", fontsize = 15, fontweight = "bold")

vald_freq.plot(kind = "pie", autopct ="%1.1f%%", ax = axes[1])
axes[1].set_title("Validation Overall Nucleotide Frequency", fontsize = 15, fontweight = "bold")

test_freq.plot(kind = "pie", autopct ="%1.1f%%", ax = axes[2])
axes[2].set_title("Test Overall Nucleotide Frequency", fontsize = 15, fontweight = "bold");

axis = [0, 1, 2]
for i in axis:
    for text in axes[i].texts:
        text.set_fontsize(16)
        text.set_family("serif")
        text.set_fontweight("bold")
```
